```python
import os
import shutil
import random

# --- CONFIGURATION ---
original_dir = 'animals'         # Your source folder
new_base_dir = 'animals_small'   # The new folder to be created

# Define how many images you want PER CLASS for Train and Test
# (Based on 1600 total images divided by 3 classes)
num_train_per_class = 427  
num_test_per_class = 107   

def create_subset():
    # 1. Clean up existing folder if it exists (Optional: safety check)
    if os.path.exists(new_base_dir):
        print(f"Warning: '{new_base_dir}' already exists. Please delete it or rename it first.")
        return

    os.makedirs(new_base_dir)

    # 2. Loop through 'train' and 'test' directories
    for split in ['train', 'test']:
        source_split_path = os.path.join(original_dir, split)
        dest_split_path = os.path.join(new_base_dir, split)
        
        # Check if source exists (e.g. do you actually have a 'test' folder?)
        if not os.path.exists(source_split_path):
            print(f"Skipping {split} (folder not found)")
            continue

        # Get list of classes (e.g., cat, dog, panda)
        classes = [d for d in os.listdir(source_split_path) if os.path.isdir(os.path.join(source_split_path, d))]
        
        for class_name in classes:
            class_src_dir = os.path.join(source_split_path, class_name)
            class_dest_dir = os.path.join(dest_split_path, class_name)
            
            # Create the specific class folder in the destination
            os.makedirs(class_dest_dir, exist_ok=True)
            
            # Get list of all images
            fnames = os.listdir(class_src_dir)
            random.shuffle(fnames) # IMPORTANT: Shuffle to get a random sample
            
            # Determine how many to copy based on the split
            limit = num_train_per_class if split == 'train' else num_test_per_class
            
            # Slice the list
            selected_fnames = fnames[:limit]
            
            print(f"[{split.upper()}] Copying {len(selected_fnames)} images for '{class_name}'...")
            
            # Copy the files
            for fname in selected_fnames:
                src_file = os.path.join(class_src_dir, fname)
                dst_file = os.path.join(class_dest_dir, fname)
                shutil.copyfile(src_file, dst_file)

    print("-" * 30)
    print(f"Success! Data created in '{new_base_dir}'")
    print(f"Use '{new_base_dir}/train' and '{new_base_dir}/test' in your code.")

# Run the function
create_subset()
```
